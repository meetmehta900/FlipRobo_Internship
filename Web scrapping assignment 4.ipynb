### Scrape the details team India’s international fixtures from bcci.tv.

```python
driver=webdriver.Edge('edgedriver_win64/msedgedriver.exe')
url='https://www.bcci.tv/'
driver.get(url)
```

```python
btn=driver.find_element_by_xpath('//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]')
btn.click()
btn = driver.find_element_by_xpath('//a[@class="navigation__link navigation__link--in-drop-down"]')
btn.click()
```

```python
urls=[]
match_page=driver.find_elements_by_xpath("//div[@class='js-list']/a")
for i in match_page:
    urls.append(i.get_attribute('href'))
```

```python
titly=[]
seriies=[]
loc=[]
daty=[]
timy=[]
for i in urls:
    driver.get(i)
    try:
        title = driver.find_element_by_xpath("//h1[@class='mc-header-info__title']/strong")
        titly.append(title.text)
    except NoSuchElementException:
        titly.append('N/A')
    
    try:
        series=driver.find_element_by_xpath("//span[@class='mc-header-info__match-description-match-type mc-theme-color']")
        seriies.append(series.text)
    except NoSuchElementException:
        seriies.append('N/A')
    
    try:
        place = driver.find_element_by_xpath("//h1[@class='mc-header-info__title']")
        loc.append(place.text.split()[-1])
    except NoSuchElementException:
        loc.append('N/A')        
    
    try:
        date = driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']/strong")
        daty.append(date.text)
    except NoSuchElementException:
        daty.append('N/A')

    try:
        timee = driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']/span")
        timy.append(timee.text)
    except NoSuchElementException:
        timy.append('N/A')
    
    time.sleep(2)
```

```python
dataframe=pd.DataFrame()
dataframe['Match_title']=titly
dataframe['Series']=seriies
dataframe['Place']=loc
dataframe['Date']=daty
dataframe['Time']=timy
dataframe.style.hide_index()
```
